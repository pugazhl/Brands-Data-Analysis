5.China’s Growing Influence – TikTok/Douyin (ByteDance) and ICBC (Industrial and Commercial Bank of China) highlight China’s rising brand power in technology and finance.

6.Germany’s Stronghold in Telecom – Deutsche Telekom is the only German brand in the top list, emphasizing Europe’s presence in telecommunications.

7.Retail & Finance are Also Key Sectors – Walmart (Retail) and ICBC (Finance) show that while tech dominates, traditional industries still hold strong brand value.

8.Significant Drop After Top Four – A noticeable decline in brand value is seen after the top four brands, indicating a large gap between tech giants and other companies.

# **Number of brands by country**

```python
a=data.groupby('country')['brand'].count().sort_values(ascending=False)
a
```

```python
plt.figure(figsize=(30,15))
# a.plot(kind='pie',x=a.index)
y=a.values
x=a.index
explode=[0.1]*len(y)
plt.pie(y,labels=x,radius=1.2,autopct='%1.2f%%',explode=explode)
plt.show()
```

# **Observations on Brand Value Distribution by Country (Pie Chart Analysis)**

1.United States Dominates – The United States holds the largest share of brand value, accounting for 38.8%, indicating a strong presence of globally recognized brands.

2.China Follows but at a Distance – China is the second-largest contributor with 14.2%, significantly lower than the U.S., but still a major player in global branding.

3.Japan, France, and Germany Hold Strong Positions – These countries contribute 6.6%, 6.4%, and 5.6%, respectively, showcasing their influence in brand value.

4.Highly Fragmented Lower Tier – Beyond the top 5, many countries have very small shares, indicating that globally dominant brands are concentrated in a few nations.

5.European Countries Show Moderate Influence – The United Kingdom, Italy, Switzerland, and Spain have small but notable contributions, reflecting strong luxury and financial brands.

6.Asian Presence is Noticeable – Besides China and Japan, countries like South Korea, India, and Singapore have a growing impact but remain far behind the U.S. and China.

7.Long Tail of Minor Contributors – Countries such as Vietnam, Luxembourg, Finland, and Sweden hold negligible brand value shares, emphasizing the dominance of a few key economies.

8.Brand Value is Concentrated – The top 3-5 countries control the majority of global brand value, highlighting economic and corporate influence in specific regions. influential markets.

# **compare top 10 companies brand values**

```python

top_10 = data.nlargest(10, 'BRAND VALUE')
rest =data[data.index.isin(top_10.index)]
plt.figure(figsize=(10, 6))
sns.histplot(data=[top_10['BRAND VALUE'], rest['BRAND VALUE']])
plt.title('Top 10 Companies vs. Others')
# plt.xticks([0, 1], ['Top 10', 'Rest'])
plt.show()
```

# **observation**
1.The histogram represents the brand values of the top 10 companies, categorized into bins.

2.Most companies fall within the lowest brand value range (around 100,000), while fewer companies are in higher value ranges (300,000–500,000).
